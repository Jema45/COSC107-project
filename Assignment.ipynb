Tuple

```python
colours = ("red", "blue", "green", "yellow", "black")
print(colours[0])
print(colours[2])
print(colours[-1])
```

```python
colours = ("red", "blue", "green", "yellow", "black")
colours = list(colours)      # convert to list
colours.append("white")      # append
colours = tuple(colours)     # back to tuple

print(colours)
```

```python
colours = ("red", "blue", "green", "yellow", "black")
colours = list(colours)
colours.pop()          # removes last item
colours = tuple(colours)

print(colours)
```

```python
colours = ("red", "blue", "green", "yellow", "black")
colours = list(colours)
colours.remove("blue")   # remove by value
colours = tuple(colours)

print(colours)
```
